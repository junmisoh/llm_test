```python
import openai
import os
import certifi
import httpx
from IPython.display import Markdown, display
# from dotenv import load_dotenv

# load_dotenv(override=True)
# amd_api_key = os.environ.get("AMD_API_KEY")

client = openai.OpenAI(
    base_url="https://llm-api.amd.com/OnPrem",
    api_key="dummy",
    http_client=httpx.Client(
        verify=certifi.where()
        ),
    default_headers={
        "Ocp-Apim-Subscription-Key": "a5ab74eb757b4b98989f6393212feb7a",
        "user": os.getlogin()
        }
)

response = client.chat.completions.create(
    model="Meta-Llama-4-Maverick-17B",
    # max_completion_tokens=200,
    temperature=0.7,
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Write an essay about love"}
    ]
)

print(response)
```
